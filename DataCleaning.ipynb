```python
import numpy as np
import pandas as pd
import time 
import datetime as datetime
import glob
import matplotlib.gridspec as gridspec
import matplotlib.pyplot as plt
np.random.seed(1)
```

```python
Full = pd.read_csv('FullData.csv')
Full.head()
```

```python
Full = Full[Full['bid price'] != 0]
Full = Full[Full['ask price'] != 0]
Full.shape
```

```python
Full.to_csv('FullData.csv',index=False)
```

```python
LP1 = Full[Full['provider'] == 'LP-1']
LP2 = Full[Full['provider'] == 'LP-2']
LP3 = Full[Full['provider'] == 'LP-3']
LP4 = Full[Full['provider'] == 'LP-4']
LP5 = Full[Full['provider'] == 'LP-5']
```

```python
print('pre drop 1', LP1.shape)
LP1['timestamp'] = LP1['time'].astype(str).str[:-4]
LP1 = LP1.drop_duplicates(['timestamp','currency pair'], 'last')
print('post drop 1', LP1.shape)
print('pre drop 2', LP2.shape)
LP2['timestamp'] = LP2['time'].astype(str).str[:-4]
LP2 = LP2.drop_duplicates(['timestamp','currency pair'], 'last')
print('post drop 2', LP2.shape)
print('pre drop 3', LP3.shape)
LP3['timestamp'] = LP3['time'].astype(str).str[:-4]
LP3 = LP3.drop_duplicates(['timestamp','currency pair'], 'last')
print('post drop 3', LP3.shape)
print('pre drop 4', LP4.shape)
LP4['timestamp'] = LP4['time'].astype(str).str[:-4]
LP4 = LP4.drop_duplicates(['timestamp','currency pair'], 'last')
print('post drop 4', LP4.shape)
print('pre drop 5', LP5.shape)
LP5['timestamp'] = LP5['time'].astype(str).str[:-4]
LP5 = LP5.drop_duplicates(['timestamp','currency pair'], 'last')
print('post drop 5', LP5.shape)
```

```python
Full_second = LP1.append(LP2).append(LP3).append(LP4).append(LP5)
```

```python
Best_column = ['time','currency price','bid provider', 'ask provider', 'bid price', 'ask price', 'bid volume', 'ask volume']
```

```python
bid_best = Full_second
ask_best = Full_second
bid_best['_ind'] = bid_best['timestamp'] + "+" + bid_best['currency pair']
ask_best['_ind'] = ask_best['timestamp'] + "+" + ask_best['currency pair']
```
